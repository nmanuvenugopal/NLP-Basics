```python
# Now we need to import the NLTK module
# Then only we can import PorterStemer and Snowball module else it will not be listed.

import nltk
```

```python
# Now import PorterStemmer and SnowballStemmer

from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
```

```python
# After importing we need to inditailize the object of both PorterStemmer and SnowballStemmer

port_stem = PorterStemmer()
snow_stem = SnowballStemmer(language='english')
```

```python
# Defining list of word for which stemming needs to be applied

words1 = ['run', 'runs', 'runner', 'running', 'ran']

words2 = ['generous', 'generously', 'generate', 'generation']

words3 = ['easy', 'easily', 'fairly', 'fairness', 'fair']
```

```python
# Applying Porter stemmer and comparing results
print("PorterStemmer")
print("---------------------------------------")

for word in words1:
    print(word + ' --------> ' + port_stem.stem(word))


print('\n')
print("SnowStemmer")
print("---------------------------------------")
for word in words1:
    print(word + ' --------> ' + snow_stem.stem(word))

```

```python
# We can see more meaningful word are generated by snowball stemmer

print("PorterStemmer")
print("---------------------------------------")

for word in words2:
    print(word + ' --------> ' + port_stem.stem(word))


print('\n')
print("SnowStemmer")
print("---------------------------------------")
for word in words2:
    print(word + ' --------> ' + snow_stem.stem(word))
```
